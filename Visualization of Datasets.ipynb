```python
#Importing Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
```

Analysing Guam Data

```python
#Reading and Displaying data
guamdata= pd.read_csv('1630000_meantrend.csv', index_col=False)
guamdata
```

```python
#Creating time step for data 
tsgd = pd.date_range('1948 1', periods=guamdata.shape[0], freq='MS')
tsgd #Display time series
```

```python
#Creating time series 
TSMSL = Series(guamdata.loc[:,' Monthly_MSL'].values, index=tsgd)
```

```python
#Plotting time series object
ax= TSMSL.plot()
ax.set_xlabel('Time')
ax.set_ylabel('Mean Sea Level')
ax.set_title('Monthly Mean Sea Level for Apra Harbor, Guam')
```

```python
plt.scatter(tsgd, guamdata[' Monthly_MSL'], marker='.', label= 'Raw Monthly Mean Sea Level')
plt.plot(tsgd, guamdata[' Linear_Trend'], color= 'red', label= 'Linear Trend of Monthly Mean Sea Level')
plt.xlabel('Time')
plt.ylabel('Mean Sea Level (m)')
plt.title('Linear Trend of MSL in Guam')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```
